# Importing an image to test with

```python
cat_test = cv2.imread('C:\\Users\\gl127\\Downloads\\yolov3_coco\\images\\cats1.jpg', cv2.IMREAD_UNCHANGED)
cat_test = cv2.resize(src = cat_test, dsize = None, fx = 0.15, fy = 0.15)
cv2.imshow('cats', cat_test)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Preprocessing an image so it can be a properly formatted input for the model

```python
blob = cv2.dnn.blobFromImage(image = cat_test,
                        scalefactor = 0.00392, # 1/sigma for normalizing
                        size = (416, 416), # seems like 609, 609 doesn't work?
                        mean = (0, 0, 0), # mean (R,G,B) values for normalizing; will be subtracted from each channel
                        swapRB = True, # need to swap red and blue because mean subtracts RGB but opencv read in BGR
                        crop = False) # don't crop when resizing

# good explanation for each parameter: https://www.pyimagesearch.com/2017/11/06/deep-learning-opencvs-blobfromimage-works/
# with mean = 0,0,0 and scalefactor = 1/255, we're just resizing the image then dividing all BGR values by 255
# so input values range in [0, 1] at the moment
```

# Pass image through NN

```python
## Need output layers to pass blobs through **************CHECK WHAT THE OUTPUTS OF EACH LOOK LIKE
```

```python
layer_names = model1.getLayerNames() # yolov3 architecture has 3 outputs at layers 82, 94, 106 labeled yolo_82, yolo_94, yolo_106
layer_names
```

```python
outlayer_place = model1.getUnconnectedOutLayers() # grab index + 1 of those output layers (i.e. 254th place is indexed 253)
outlayer_place
```

```python
output_layers = [layer_names[i[0]-1] for i in outlayer_place] # need -1 to get index
output_layers
```

```python
model1.setInput(blob) # "sets the new input value for the network"
```

```python
preds = model1.forward(output_layers) # passes input through network to get output, they call the output "output blobs"
```

# Get info from outputs

```python
## Need to grab size/shape of image because the neural network outputs normalized bounding box dimensions, i.e. [0,1]
```

```python
cat_test.shape # (490, 367, 3)
```

```python
height = cat_test.shape[0]
width = cat_test.shape[1]
n_channels = cat_test.shape[2] # remember opencv reads images in BGR, not RGB or BGRa
```

```python
boxes = []
class_probs_list = []
class_ids = []

for pred in preds: 
    for detect_loc_probs in pred: # each detect_loc_probs contains 0: normalized center x-coord of detected obj's bounding box, 
                                # 1: normalized center y-coord of detected obj's bounding box, 2: normalized width of detected obj's bounding box,
                                # 3: normalized height of detected obj's bounding box
        class_probs = detect_loc_probs[5:] # probabilities of being each class (out of the 80 classes), which are the last 80 values
        class_idx = np.argmax(class_probs) # what is it most likely to be? -> grab that index to find 
        max_prob = class_probs[class_idx] # the highest probability
        
        if max_prob > 0.5: # typically 0.5
            center_x = int(detect_loc_probs[0] * width) # detect_loc_probs[0] are numpy.float32s
            center_y = int(detect_loc_probs[1] * height) # grabbing "rescaled" lengths back
            w = int(detect_loc_probs[2] * width) # center_x, center_y, w, and h can define a bounding box but cv.rectangle uses top left corner to bottom right
            h = int(detect_loc_probs[3] * height)
            top_left_x = center_x - w/2
            top_left_y = center_y - h/2
            
            # keep track of instances/bounding boxes where there is likely to be an object
            boxes.append([w, h, top_left_x, top_left_y])
            # keep track of that instance's probability and class
            class_probs_list.append(max_prob)
            class_ids.append(class_idx)
            
```
